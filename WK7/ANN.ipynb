```python
import tensorflow as tf
import pandas as pd
```

```python
rf=pd.read_csv("Churn_Modelling.csv")
```

```python
rf.head(5)
```

```python
fdata=rf.drop(['RowNumber'],axis=1)
```

```python
fdata=rf.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
fdata.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
label_geo=LabelEncoder()
label_Gender=LabelEncoder()
```

```python
fdata['Geography']=label_geo.fit_transform(fdata['Geography'])
```

```python
fdata.head(4)
```

```python
fdata['Gender']=label_Gender.fit_transform(fdata['Gender'])
```

```python
fdata.head(4)
```

```python
sc=StandardScaler()

x=sc.fit_transform(fdata.drop('Exited',axis=1))
```

```python
y=fdata['Exited']
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python


# Input layer + 1st hidden layer
model.add(Dense(16, activation='relu', input_shape=(x.shape[1],)))

# 2nd hidden layer
model.add(Dense(8, activation='relu'))

# Output layer (binary classification → sigmoid)
model.add(Dense(1, activation='sigmoid'))

# 2. Model compile karein
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 3. Model train karein
history = model.fit(x, y, epochs=20, batch_size=32, validation_split=0.2)
```

```python
loss, acc = model.evaluate(x, y)
print(f"Training Accuracy: {acc*100:.2f}%")
```

```python
y_pred = model.predict(x)
y_pred_class = (y_pred > 0.5).astype(int)
```
